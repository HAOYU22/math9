# Conditional statements

Suppose we wanted to define a function in a piecewise manner, e.g. 

$$f(x) =
\left\{
	\begin{array}{ll}
		0  & \mbox{if } x < 0 \\
                x & \mbox{otherwise }
	\end{array}
\right.$$

This function is called the RELU function (RE for 'rectified' and LU for 'linear unit').

We can do this in Python (or any programming language) via the **if-else** syntax:


````python
def relu(x):
    if x > 0:  # this is a comparison (more on this later)
        return x
    else:
        return 0.0   # bug: np.vectorize in math9.graphfunction gets confused if we put 0 instead of 0.0

math9.graphfunction(relu)
````

Actually, we could have written it like this:

````python
def relu(x):
    if x > 0:
        return x
    return 0.0    # no need for else, because return will end the function anyway

math9.graphfunction(relu)
````

Another example where you don't need to have an `else` where there is an `if`:

````python
def safe_div(x, y):
    if y == 0:
        print("you are dividing by zero!")
        return x/0.0000000000001
    return x/y

safe_div(1,0)
````

Recall this `if-else` construction:

````python
if True: 
    x = 0
else: 
    x = 1
x
````

We can write this more succinctly via an inline `if-else` syntax: 

````python
x = 0 if True else 1
print x
````

````python
x = 0 if False else 1
print x
````

So there is a type called `bool`. It can have values `True` or `False`.

Let's code up the following function: 
$$f(x) =
\left\{
	\begin{array}{ll}
        0.5  & \mbox{if } 0 \leq x \leq 0.5 \\
        0 & \mbox{otherwise }
	\end{array}
\right.$$




````python
def f(x):
    if x >= 0:
        if x <= 0.5:
            return 0.5
    return 0.0
````

````python
math9.graphfunction(f)
````

Why not combine `x >= 0` and `x <= 0.5`:

````python
def f(x):
    if x >= 0 and x <= 0.5:
            return 0.5
    return 0.0
````

````python
math9.graphfunction(f)
````
