### Problem 4

Functions of functions. 
Functions are just like any other variable and can be returned from functions
Write down a function called `rotateleft` that will take a function f and will rotate its picture. 

```python
def rotateleft(f):
    def g(x,y):
        return f(y, -x)
    return g
```

```python
def fun(x,y):
    return abs(x+y)
hwlib.drawfunction(fun)
```

```python
hwlib.drawfunction(rotateleft(fun))
```

```python
def repeat(f):
    def g(x,y):
        if x<0:
            newx = 2*(x+0.5)
        else:
            newx = 2*(x-0.5)
        if y<0:
            newy = 2*(y+0.5)
        else:
            newy = 2*(y-0.5)
        return f(newx,newy)
    return g
```
